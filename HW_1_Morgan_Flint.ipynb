## a)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,500)
g0 = np.exp(-np.sqrt(2)*np.abs(x))/np.sqrt(2)
g1 = np.exp(-((x)**2)/2)/np.sqrt(2*np.pi)

plt.figure()
plt.plot(x,g0,label = "G0")
plt.plot(x,g1,label = "G1")
plt.legend()
plt.show()
```

## b)

```python
x = np.linspace(-5,5.525,500)
g0 = np.exp(-np.sqrt(2)*np.abs(x))/np.sqrt(2)
g1 = np.exp(-((x)**2)/2)/np.sqrt(2*np.pi)
h_star = np.maximum(g0,g1)
plt.figure()
plt.plot(x,h_star,label = "h*")

plt.legend()
plt.show()
```
