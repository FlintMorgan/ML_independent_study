```python
import numpy as np
import matplotlib.pyplot as plt
n = 10;m=2;p_i = 0.5
E = np.linspace(0,1,500)

P = 2*m*np.exp(-2*(E**2)*n)

hi = 2*np.exp(-2*n*E**2)

plt.figure()
plt.plot(E,P,label = "P")
plt.plot(E,hi,label = "huffing ineq")
plt.legend()
plt.show()
```

## a)

```python


x = np.linspace(-5,5,500)
g0 = np.exp(-np.sqrt(2)*np.abs(x))/np.sqrt(2)
g1 = np.exp(-((x)**2)/2)/np.sqrt(2*np.pi)

plt.figure()
plt.plot(x,g0,label = "G0")
plt.plot(x,g1,label = "G1")
plt.legend()
plt.show()
```

## b)

```python
x = np.linspace(-5,5.525,500)
g0 = np.exp(-np.sqrt(2)*np.abs(x))/np.sqrt(2)
g1 = np.exp(-((x)**2)/2)/np.sqrt(2*np.pi)
h_star = np.maximum(g0,g1)
plt.figure()
plt.plot(x,h_star,label = "h*")

plt.legend()
plt.show()
```
